# Matplotlib and Data Visualization
Data visualization is one of the most important components in data analytics and is a great tool for data scientists. It allows us to accually see the data in an easier, more understandable format than blankly starring at the data. Ranges and parameters are useful, but they do not give us the information about all the data, while if we visualize it, we can sometimes see the patterns in the data, and even spot the flaws, like outliers or NaN values. 

In Python we have quite a few liblaries dedicated to data visualization. In this notes, we will work on *Matplotlib*, which is the most popular and common one. It allows for creation of any desired chart or plot and it is relatively easy to use. While very useful, the biggest drawback for Matplotlib is that it might be hard to create complex visualization using just this liblary. We will frequently use it for analysis, but for final reports or visualizations for clients, other liblaries might provide better looking option. Still, Matplotlib is a great tool and many other visualization liblaries work similarely or even on top of it, therefore grasping a good understanding of the basics will come in handy.

Other liblaries that I will be interested in exploring:
- Seaborn (built on top of Matplotlib, has prettier visualization with less syntax, but also has narrower options);
- Plotly (allows for interactive visualizations, also quite pretty);
- Bokeh (also allows for interactive visualizations, has similar syntax to Matplotlib);
- Folium (is relatively easy and is solemnly dedicated to geospatial visualizations);
- Panel (allows for creation of pretty plots and easy, interactive dashboards)

## 1. Basic plot and it's elements
Visualizations help us with understaning the data. Since there are a lot of different types of data, different statistics and different methods, hence there are quite a lot of visualization types. In here we will generally refer to them as graphs or plots or call them by their distingt name. There no one-size-fits-all type of plot - however, in matplotlib all the plot types follow almost the same syntax when it comes to creating the whole visualization and most of the time, the method name and logic behind assigning the values to axes will create a whole lot of difference.

In this part of the notes we will go through all the elements of the graph, how to change these elements and we'll add a few more functions to wrap it up.

### Creating a graph
We have two ways of creating a graph in matplotlib.

For the easier one, we will call upon `plt.plot(x, y)` and this should do the trick. After that, we will define other elements. If we are working with data and we just need a quick look at the histogram or the distribution or we want to spot the outliers - this is a totally fine option and will give us all the basics we might need.

However, if we want to create a slightly more advanced visualization and we do not want to define the figure separately, we will use the following syntax:
```python
fig, ax = plt.subplots()
ax.plot(x, y)
```
Although it is longer at this point, the more we add to the graph, the more syntax will we safe and it's nice and packed in one place. It is also easier to plot more kinds of different data using this syntax, rather than the simple one. I will be focusing on using this one in these notes, but I will certainly use both methods in my future analysis.

````python
# let's plot some data in both ways
plt.plot(df['Time'], df['Sales'])
````

````python
# And the same one the other way
fig, ax = plt.subplots()
ax.plot(df['Time'], df['Sales'])
````

The plot is created the exact same way using both ways. More on the difference can be read in [here](https://towardsdatascience.com/clearing-the-confusion-once-and-for-all-fig-ax-plt-subplots-b122bb7783ca). 

The other difference between these two ways of creating a graph is how we will assign all the additional elements in these. It's not humongous difference, but it needs to be noted, to not cause any trouble in the future. 



### Title, labels and text formatting
The plots we created above are not giving a lot of informations - while we know what they stand for, since we created them, if we show them to any random person, they will be very much confused. A good graph needs it's labels and preferably a title, so we know what we are looking at.

Within all the text element in the graph, besides deciding on the description, we can also change the properties of our text with `fontdict{}`. Within the fontdict, we can format the size, color, rotation, placement, font (font doesn't work in Google Colaboratory this way), etc. For some text elements, like ticks labels there are also different formatting methods and we will try one later on.


````python
# let's use the basic syntax one more time and see how we create title and labels here
plt.plot(df['Time'], df['Sales'])

plt.title('Sales through time')
plt.xlabel('Time')
plt.ylabel('Sales')

# with the basic syntax we use these methods
````

````python
# when using fig, ax, we will be using setter methods
fig, ax = plt.subplots()
ax.plot(df['Time'], df['Sales'])

#let's play with the font a little in the title
ax.set_title('Sales through time', pad=20, fontdict={'size':20, 
                                             'rotation': 180, 
                                             'color': 'purple',
                                             'family': 'fantasy',
                                             'ha': 'left'})
# using pad= gives us the space from the plot in the title

ax.set_xlabel('Time', loc='right') # loc= let's us decide where we place the label
ax.set_ylabel('Sales', labelpad=30) # for labels we need to use labelpad=
# I know it's not so pretty, but to you see these options? Cool, right?
````

### Lines and markers
Another element we cannot skip in the visualization, the key element, is our data. In many different visualizations we will show it in different way, let's focus on lines and markers, as there will appear frequently and if we get them, all other patterns will be easy to follow.

Using the fig, ax notation, we can add as many additional lines or points, etc., as we want - of course our graph needs to be readable and make sense, and also it is not a good practice to add too much data, however this is possible. It is also possible with the standard notation.

Our lines will follow the values we add to the graph and connect them, while if we use the markers, these will only show us the values. Depending on what we want to show, we can use either or one of these or they can show us some interesting dependencies.


````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

# for now, let's simply display the data
ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])
````

````python
# Alright, let's do dome visualization magic
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')


# there are two ways to change the visual of our lines or points
# well, it's actually one way, but one is descriptive
ax.plot(df['Time'], df['Sales'], color='#573950', marker='o', ls='dashed', dash_capstyle='round', linewidth=4, markersize=10)

# or we can used shorthand notation which goes: "ColorMarkerLinestyle" (we can skip any of these, but the order is important)
ax.plot(df['Time'], df['Sales2'], 'C6p:', linewidth=4, markersize=10) 
````

````python
# now let's try with points
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

# when working with any different type of data visualization, if we want to change it
# the best pratice is to just check the documentation - we have all keys there
ax.scatter(df['Time'], df['Sales'], marker='$$', color='pink', edgecolor='purple', s=50)
ax.scatter(df['Time'], df['Sales2'], marker='*', color='yellow', edgecolor='black', s=50)
#cute, ain't it?
````

We can do as many things with these visualisations as we want. 

We will also work on another example, where we will split the line, but this way it can be used on most types of plots.

There are also other possibilities - we can create a x**2 line, we could also filter the plot or scatters so that values above or below some value will have different color or shape, but fot now I will skip it and learn if it comes up.

````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

# if we would like to split the data as one point, we need to specifi the position and create two plots
ax.plot(df['Time'][:4], df['Sales'][:4], 'r')
ax.plot(df['Time'][3:], df['Sales'][3:], 'b--')
````

### Creating labels on the plot or scatters
Since now we know how to create plots and markers properly, sometimes the data on the plot might not look too precise - if we create a small graph for the sake of saving space in a document, we might end up with not so clear visualization. This is where the labels on the plot come in handy - if we cannot clearly see what is the value, we can just print it.

````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])
# if we want to loop through all the values, it is important to.. use a loop indeed
for i,j in zip(df['Time'],df['Sales']):
    ax.annotate(str(j),xy=(i,j))

# and let's say we want to display a different label, from different column
for i,j,k in zip(df['Time'],df['Sales2'],df['DT']):
    # and if it's not clear to read, we can put the label in a box or offset it (we will do offset another time)
    ax.annotate(str(k),
                xy=(i,j),
                bbox={'boxstyle': 'ellipse', 'fc':'pink', 'alpha': 0.7})
````

### Ticks and ticklines
Continuing the topic of text elements, we have to mention ticklabels and therefore we need to mention ticks.

Both x- and yticks will be automaticly generated and will use the min and the max value to define their ranges - it is a nice information to have to spot the outliers. If most of our data spreads is visible in one place and there is no data visible but a very long x or y axis, it says that there might be some outliers on this value.

Also important to know is that the graph will not allow us to create the axes larger than generated - if we would like to extend the yaxis in the above example to like, 2000, then we would need a value close to it in the plotted data.

````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])

ax.minorticks_on
ax.set_xticks(df['Time'])
ax.set_xticklabels(['a','b','c','d','e','f'])
````

### Legend
The more we plot on the graph or the more colours we use, we must explain what we are showing on there, so anyone could understand or reasoning for these. With the legend, we can quickly explain, what is shown on the plot and how it is presented.

Within the legend, we can edit everything - text, title, spacing and size.

````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'], color='#573950', marker='o', ls='dashed', dash_capstyle='round', linewidth=4, markersize=10)
ax.plot(df['Time'], df['Sales2'], 'C6p:', linewidth=4, markersize=10) 

ax.legend(['Sales', 'Sales2'], labelcolor='orange', title='Legend')
````

### Grid
Grid is another element, that could tie our visualization together and make it more readable. Especially when we have many values and labels will not make us much help, grid can navigate the reader and show the exactly where the value lays, or at least know closer. 

````python
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'], color='#573950', marker='o', ls='dashed', dash_capstyle='round', linewidth=4, markersize=10)
ax.plot(df['Time'], df['Sales2'], 'C6p:', linewidth=4, markersize=10) 

ax.legend(['Sales', 'Sales2'], labelcolor='orange', title='Legend')
ax.grid()
````

### Changing the default style
Matplotlib has one default style, which can be seen above. However, there are ways to make the graph prettier or different quite easily. Besides editing all option and colors one by one, we can also use some of the available motives and apply them to all our graphs.

**Creating our own style**   
Another fun option when creating more visualizations in matplotlib is that we can create our own styles.   
To do that, we need to create a new file with the extention `.mplstyle` and then provide the source to it, if it's not in the same folder as the code.
Then we can either write the setting ourself or copy from either existing mpl style or from other styles available online, 
and if we would like to ajust the colors or sizes, we can do it all in that file and then have all of our plots made similar.

````python
plt.style.use('dark_background') # if we set this at the beginning or at one point, 
# all following graphs will be in this style
fig, ax = plt.subplots()

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])

ax.legend(['Sales', 'Sales2'], labelcolor='orange', title='Legend')
ax.grid()
````

### Changing the figure
The last thing to edit will be the figure.

We can enlarge or minimizethe graph in the figure, but also set up a layout in here. We will talk about in a bit.



````python
plt.style.use('default') # we change to default
fig, ax = plt.subplots(figsize=(20,10), facecolor='yellow', dpi=200)

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])

ax.legend(['Sales', 'Sales2'], labelcolor='blue', title='Legend')
ax.grid()
````

### Cleaner display and saving the image
The last thing to do with our beautiful visualizations is to display them properly, so that we don't get that traceback above. 

If we will be using another tool for the report, it would be also best to save the plot as png or jpg, so that it could be in the best quality, instead of making a screenshoot.

````python
fig, ax = plt.subplots(figsize=(20,10), facecolor='yellow', dpi=200)

ax.set_title('Sales through time', pad=20)
ax.set_xlabel('Time')
ax.set_ylabel('Sales')

ax.plot(df['Time'], df['Sales'])
ax.scatter(df['Time'], df['Sales2'])

ax.legend(['Sales', 'Sales2'], labelcolor='blue', title='Legend')
ax.grid()

plt.savefig('Sales_time.png') #this will save our visualization in the folder
plt.show() # and this how we display it
````

## 2. Creating a dashboard in matplotlib
Another commonly seen visualizations are dashboards - these allow us to show more than one plot at once in a neat frame. 

What is important in matplotlib, when creating a dashboard is that it is a bit harder to encode the scalability than in other python libraries (it is possible though). For this example, our dahsbord will be static.

Another thing to mention is that for the dashboards we need to import another sublibrary from matplotlib, which is **Gridspec**. It will create the grid in which we will encode our plots.

Also, differently than with the other plots, we will now need to enumerate the plots we create - with the numerations we will assign the positions in the grid.

````python
titanic
````

````python
import matplotlib.gridspec as gridspec
````

````python
# for this example we will use the titanic dataset
with plt.style.context(("seaborn","ggplot")):
  fig = plt.figure(constrained_layout=True, figsize=(10,8)) # with constrained_layout the layout will be
# automaticly corrected
  specs = gridspec.GridSpec(ncols=2, nrows=2, figure=fig) #here we define the shape of the dashboard

  ax1 = fig.add_subplot(specs[0, :]) ## First Row
  ax2 = fig.add_subplot(specs[1, 0]) ## Second Row First Column
  ax3 = fig.add_subplot(specs[1, 1]) ## Second Row Second Colums

# for the first plot let's create a scatter plot for the fare and classes
  color = {'C': 'tab:red', 'Q': 'tab:blue', 'S':'tab:orange', 'NaN': 'tab:gree'}
  for embarke in titanic['Embarked'].dropna().unique():
    ax1.scatter(titanic[titanic['Embarked']==embarke]['Age'],
                titanic[titanic['Embarked']==embarke]['Fare'],
                c=color[embarke],
                s=100,
                label=embarke)
    ax1.set_xlabel('Age')
    ax1.set_ylabel('Fare')
    ax1.set_title('Age vs. Fare for each embarke')
    ax1.legend(title='Embarkes', loc='best')

# for the second plot let's create a pie chart with all the sex and if they survived or not
  sex_surv = titanic.groupby(['Sex', 'Survived']).count()[['PassengerId']]
  sex_surv.reset_index()
  sex_survival = sex_surv.groupby('Sex')['PassengerId'].apply(list).to_dict()
  labels = list(sex_survival.keys())
  values = list(sex_survival.values())
  values = np.array(values)

  cmap = plt.colormaps["tab20c"]
  outer_colors = cmap(np.arange(3)*4)
  inner_colors = cmap([1, 2, 5, 6, 9, 10])
  size = 0.3

  ax2.pie(values.sum(axis=1), labels=labels, radius=1, 
        colors=outer_colors,
        wedgeprops=dict(width=size, edgecolor='w'))

  ax2.pie(values.flatten(), radius=0.7, colors=inner_colors,
        wedgeprops=dict(width=size, edgecolor='w'))
  
  ax2.set_title('Passengers sex and surival percentages')

# for the last one let's do histogram of age
  ax3.hist(titanic['Age'], bins=8)
  ax3.set_xlabel('Passengers age')
  ax3.set_ylabel('Passengers count')
  ax3.set_title('Passengers ages')


plt.savefig('titanic_layout.png')
````
